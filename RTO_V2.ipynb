```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set() # Apply the default seaborn theme, scaling, and color palette

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("RTOdata.csv")
df2 = pd.read_csv("RTO_data_new.csv")
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
del df['OrderId']
```

```python
df['OrderPlacedDay'] = pd.to_datetime(df['OrderPlacedDay'] )
df['OrderConfirmDayOverPhone'] = pd.to_datetime(df['OrderConfirmDayOverPhone'] )
df.head()
```

```python
df['DelayConfirmation'] =df['OrderConfirmDayOverPhone'].dt.date - df['OrderPlacedDay'].dt.date
df.head()
```

```python
for ind in df.index:
    df['DelayConfirmation'][ind] = str(df['DelayConfirmation'][ind]).replace('00:00:00','').replace('days','')
    df['DelayConfirmation'][ind] = int(df['DelayConfirmation'][ind])
df.head()
```

```python
df['OrderDate'] = df.OrderPlacedDay.dt.dayofweek
df['ConfirmDate'] = df.OrderConfirmDayOverPhone.dt.dayofweek
df.head(2)
```

```python
del df['OrderConfirmDayOverPhone']
del df['OrderPlacedDay']
```

```python
df.tail(2)
```

```python
df.isna().sum()
```

```python
df.columns = ['UserId','OrderValue','PaymentType','District','OrderSource','OrderType','DeliveryCharge','IsCartOrder','OrderFromPromotionalEvent','isRTO'
             , 'Courier','DelayConfirmation','OrderDate','ConfirmDate'
             ]
```

```python
for ind in df.index:
    if pd.isnull(df['OrderType'][ind]):
        if df['DeliveryCharge'][ind] >=75:
            df['OrderType'][ind] = 'express'
        else:
            df['OrderType'][ind]='normal'
```

```python
df.isna().sum()
```

```python
# data = df.copy()
# del df['courierService']
```

```python
df['Courier'].fillna(df['Courier'].value_counts().idxmax(),inplace=True)
df['DeliveryCharge'].fillna(55,inplace=True)
```

```python
cols = ['Courier','OrderType','OrderSource','PaymentType']
for col in cols:
    print(df[col].unique())
```

```python
df.isna().sum()
```

```python
cat_cols = ['PaymentType','District','OrderSource','OrderType','IsCartOrder','OrderFromPromotionalEvent','isRTO','Courier']
for col in cat_cols:
    df[col] = df[col].astype('category')
    print(col,'---->', dict(enumerate(df[col].cat.categories)))
    df[col] = df[col].cat.codes
```

```python
df.head(2)
```

```python
df.corr()
```

```python
df['Courier'].unique()
```

```python
del df['Courier']
df.DelayConfirmation = df.DelayConfirmation.astype('int64')
```

```python
from sklearn.model_selection import train_test_split

X = df.loc[:, df.columns!='isRTO']
y = df.loc[:, df.columns=='isRTO']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

print("""X_train:{}
y_train:{}
X_test: {}
y_test: {}""".format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

model1 = LogisticRegression()
model1.fit(X_train, y_train)
y1_pred = model1.predict(X_test)
acc1 = accuracy_score(y_test, y1_pred)

model2 = LogisticRegression(C=0.05179474679231213, penalty='l2')
model2.fit(X_train, y_train)
y2_pred = model2.predict(X_test)
acc2 = accuracy_score(y_test, y2_pred)

print("""Model 1: {}
Model 2: {}""".format(acc1, acc2))
```
